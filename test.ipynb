```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV data
df = pd.read_csv('results/all_models.csv')

# Keep only the necessary columns
df_subset = df[['Model-Name', 'Modality', 'Sensitivity', 'Specificity', 'Precision']]

# Filter rows where the Modality is "['All']"
df_all = df_subset[df_subset['Modality'].str.contains("VascularDiagnosis")]

# Aggregate metrics by Model-Name (using the mean, if there are multiple rows per model)
df_grouped = df_all.groupby('Model-Name')[['Sensitivity', 'Specificity', 'Precision']].mean().reset_index()

# Create a grouped bar plot
x = np.arange(len(df_grouped['Model-Name']))  # label locations for each model
width = 0.25  # width of each bar

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, df_grouped['Sensitivity'], width, label='Sensitivity')
rects2 = ax.bar(x, df_grouped['Specificity'], width, label='Specificity')
rects3 = ax.bar(x + width, df_grouped['Precision'], width, label='Precision')

# Labeling the plot
ax.set_ylabel('Metric Value')
ax.set_title('Comparison of Model Metrics by Model-Name (Modality: VascularDiagnosis)')
ax.set_xticks(x)
ax.set_xticklabels(df_grouped['Model-Name'], rotation=45, ha='right')
ax.legend()

# Set more granular y-axis ticks (e.g., 0.85, 0.86, 0.87, ...)
ax.set_yticks(np.arange(0.00, 1.01, 0.05))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV data
df = pd.read_csv('results/all_models.csv')

# Keep only the necessary columns
df_subset = df[['Model-Name', 'Modality', 'Sensitivity', 'Specificity', 'Precision']]

# Filter rows where the Modality is "['All']"
df_all = df_subset[df_subset['Modality'].str.contains("All")]

# Aggregate metrics by Model-Name (using the mean, if there are multiple rows per model)
df_grouped = df_all.groupby('Model-Name')[['Sensitivity', 'Specificity', 'Precision']].mean().reset_index()

# Create a grouped bar plot
x = np.arange(len(df_grouped['Model-Name']))  # label locations for each model
width = 0.25  # width of each bar

# Generate random colors for each metric
color1 = np.random.rand(3,)  # random color for Sensitivity
color2 = np.random.rand(3,)  # random color for Specificity
color3 = np.random.rand(3,)  # random color for Precision

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, df_grouped['Sensitivity'], width, label='Sensitivity', color=color1)
rects2 = ax.bar(x, df_grouped['Specificity'], width, label='Specificity', color=color2)
rects3 = ax.bar(x + width, df_grouped['Precision'], width, label='Precision', color=color3)

# Labeling the plot
ax.set_ylabel('Metric Value')
ax.set_title('Comparison of Model Metrics by Model-Name (Modality: All)')
ax.set_xticks(x)
ax.set_xticklabels(df_grouped['Model-Name'], rotation=45, ha='right')
ax.legend()

# Set more granular y-axis ticks (e.g., 0.85, 0.86, 0.87, ...)
ax.set_yticks(np.arange(0.00, 1.01, 0.05))

plt.tight_layout()
plt.show()
```
